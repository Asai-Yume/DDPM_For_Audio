```python
import torch
import matplotlib.pyplot as plt
import math


lmbda = torch.arange(200, dtype=torch.float32)
lmbda.apply_(lambda x: (((0.3704*math.e)**(0.5*x))-1)**4) #set lmbda schedule
lmbda[-1] = 1
x = torch.arange(200)

plt.plot(x, lmbda)
plt.xticks(list(reversed([200, 175, 150, 125, 100, 75, 50, 25, 0])),[200, 175, 150, 125, 100, 75, 50, 25, 0])
plt.title('lambda schedule')
plt.legend()
```
