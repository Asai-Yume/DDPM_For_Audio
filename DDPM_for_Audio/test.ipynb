```python
from IPython.display import Audio, display
import matplotlib.pyplot as plt
import numpy as np
import sampling
from utils import display_waveform, save_as_ani
import librosa
import torch
import soundfile

api_ = {"model_dir": "/Users/roywang/Schools/Project/DDPM_for_Audio/third_run"}

model, device = sampling.setup(api_)
sample_rate = model.params.sample_rate
```

```python
audio = sampling.sample(model, device)
audio = audio.squeeze(0).detach().cpu()
display(Audio(audio,rate=sample_rate))
```

```python
soundfile.write("base_3.wav", audio, sample_rate)
```

```python
new_wav = sampling.infer(model, sampling.interpolation_extention, "/Users/roywang/Schools/Project/DDPM_for_Audio/base_3.wav", device)
new_wav = new_wav.squeeze(0).detach().cpu()
print(new_wav.shape)
#300k steps model
soundfile.write("audios/interpolation_extend_5_v2_sec.wav", new_wav, sample_rate)
```

```python

new_wav_2 = sampling.infer(model, sampling.normal_ext, "/Users/roywang/Schools/Project/DDPM_for_Audio/base_3.wav", device)
new_wav_2 = new_wav_2.squeeze(0).detach().cpu()
print(new_wav_2.shape)
#300k steps model
soundfile.write("audios/resampling_extend_5_sec.wav", new_wav_2, sample_rate)
```

```python
new_wav_3 = sampling.infer(model, sampling.resampling_extension, "/Users/roywang/Schools/Project/DDPM_for_Audio/base_3.wav", device)
new_wav_3 = new_wav_3.squeeze(0).detach().cpu()
print(new_wav_3.shape)
#300k steps model
soundfile.write("audios/resampling_real_extend_5_sec.wav", new_wav_3, sample_rate)
```

```python
reference, _ = librosa.load("/Users/roywang/Schools/Project/DDPM_for_Audio/base_3.wav")
display_waveform(reference,text='Original', sample_rate=sample_rate)
display(Audio(reference,rate=sample_rate))
reference = torch.tensor(reference.astype(np.float32)).unsqueeze(0).to('mps')
```

```python
new_wav, tgt_th_time = sampling.interpolation_extention(model, reference[:, :88200], device, track_history=True)
new_wav = new_wav.squeeze(0).detach().cpu()
print(new_wav.shape)
#300k steps model
save_as_ani(tgt_th_time, 'interpolation', '/Users/roywang/Schools/Project/DDPM_for_Audio/images')
display_waveform(new_wav,text='interpolation', sample_rate=sample_rate)
display(Audio(new_wav,rate=sample_rate))
```

```python
new_wav_2, tgt_th_time = sampling.normal_ext(model, reference[:, :88200], device, track_history=True)
new_wav_2 = new_wav_2.squeeze(0).detach().cpu()
print(new_wav_2.shape)
#300k steps model
save_as_ani(tgt_th_time, 'normal_concat', '/Users/roywang/Schools/Project/DDPM_for_Audio/images')
display_waveform(new_wav_2,text='resampling', sample_rate=sample_rate)
display(Audio(new_wav_2,rate=sample_rate))
```

```python
new_wav_3, tgt_th_time = sampling.resampling_extension(model, reference[:, :88200], device, track_history=True)
new_wav_3 = new_wav_3.squeeze(0).detach().cpu()
print(new_wav_3.shape)
#300k steps model
print(len(tgt_th_time))
save_as_ani(tgt_th_time, 'repaint_concat', '/Users/roywang/Schools/Project/DDPM_for_Audio/images')
display_waveform(new_wav_3,text='resampling', sample_rate=sample_rate)
display(Audio(new_wav_3,rate=sample_rate))
```
